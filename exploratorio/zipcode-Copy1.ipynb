```python
import pandas as pd
import time
from locationiq.geocoder import LocationIQ
modelar = pd.read_csv("modelardefinitivo.csv")
#estimar = pd.read_csv("estimardifinitivo.csv")
modelar = modelar.reset_index()
modelar = modelar.drop(columns="index")
modelar = modelar.drop(columns="Unnamed: 0")
#estimar = estimar.drop(columns="Unnamed: 0")
```

```python
#PONER VUESTRA API KEY
#OS REGISTRAIS AQUI: https://locationiq.com/register
#una vez registrado os lleva directamente a una pagina donde os pone vuestra API key
geocoder = LocationIQ("bfbe3f31158dd4")
```

```python
modelar["place_id"]=""
modelar["license"]=""
modelar["lat"]=""
modelar["lon"]=""
modelar["display_name"]=""
modelar["boundingbox1"]=""
modelar["boundingbox2"]=""
modelar["boundingbox3"]=""
modelar["boundingbox4"]=""
modelar["importance"]=1.1
modelar["house_number"]=""
modelar["road"]=""
modelar["neighbourhood"]=""
modelar["city"]=""
modelar["state"]=""
modelar["country"]=""
modelar["postcode"]=""
modelar["country_code"]=""
modelar.columns
modelar
```

```python
###ESTA ES LA CELDA QUE TENEIS QUE DEJAR EJECUTANDO.
##CAMBIAR EL RANGO CADA UNO SEGUN LO QUE OS TOQUE. 
#MANU DEL 30000 AL 65000 
#PABLO DEL 65001 AL 95001
#HAY QUE HACER 10000 CADA DÍA SI ALGUIEN ESTA DISPUESTO A HACERSE OTRA CUENTA EN EL PORTATIL DEL CURRO LO PODEMOS HACER
#MAS RAPIDO
for fila in range(85000,95000): 
    datos = geocoder.reverse_geocode(modelar.iloc[fila,57],modelar.iloc[fila,56])   
    try:
        modelar.iloc[fila,58]=datos["place_id"]
    except:
        modelar.iloc[fila,58]="no disponible"
    try:
        modelar.iloc[fila,59]=datos["licence"]
    except:
        modelar.iloc[fila,59]="no disponible"
    try:
        modelar.iloc[fila,60]=datos["lat"]
    except:    
        modelar.iloc[fila,60]="no disponible"
    try:
        modelar.iloc[fila,61]=datos["lon"]
    except:
        modelar.iloc[fila,61]="no disponible"
    try:
        modelar.iloc[fila,62]=datos["display_name"]
    except:
        modelar.iloc[fila,62]="no disponible"
    try:
        modelar.iloc[fila,63]=datos["boundingbox"][0]
    except:
        modelar.iloc[fila,63]="no disponible"
    try:
        modelar.iloc[fila,64]=datos["boundingbox"][1]
    except:    
        modelar.iloc[fila,64]="no disponible"
    try:
        modelar.iloc[fila,65]=datos["boundingbox"][2]
    except:
        modelar.iloc[fila,65]="no disponible"
    try:
        modelar.iloc[fila,66]=datos["boundingbox"][3]
    except:
        modelar.iloc[fila,66]="no disponible"
    try:
        modelar.iloc[fila,67]=datos["importance"]
    except:
        modelar.iloc[fila,67]="no disponible"
    try:
        modelar.iloc[fila,68]=datos["address"]["house_number"]
    except:   
        modelar.iloc[fila,68]="no disponible"
    try:
        modelar.iloc[fila,69]=datos["address"]["road"]
    except:
        modelar.iloc[fila,69]="no disponible"
    try:
        modelar.iloc[fila,70]=datos["address"]["neighbourhood"]
    except:
        modelar.iloc[fila,70]="no disponible"
    try:
        modelar.iloc[fila,71]=datos["address"]["city"]
    except:
        modelar.iloc[fila,71]="no disponible"
    try:
        modelar.iloc[fila,72]=datos["address"]["state"]
    except:    
        modelar.iloc[fila,72]="no disponible"
    try:    
        modelar.iloc[fila,73]=datos["address"]["country"]
    except:    
        modelar.iloc[fila,73]="no disponible"
    try:    
        modelar.iloc[fila,74]=datos["address"]["postcode"]
    except:
        modelar.iloc[fila,74]="no disponible"
    try:
        modelar.iloc[fila,75]=datos["address"]["country_code"]
    except:
        modelar.iloc[fila,75]="no disponible"
    time.sleep(1)
modelar.head()
```
